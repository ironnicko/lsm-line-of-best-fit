```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
"""so to find your prediction, just get the download link from Yahoo Finanace
of your favui

"""
csv = ""
df = pd.read_csv(csv)
```

```python
main = df[["Close", "Date"]]
y = main["Close"]
y.fillna(method='ffill', inplace=True)
x = main["Date"]
main.insert(0, "Days", range(len(df)))
```

```python
dimension = np.polyfit(main["Days"], y, 1)
frame = np.poly1d(dimension)
```

```python
main.insert(0,'Line of Best-Fit',frame(main['Days']))
```

```python
ax = main.plot(x="Days", y="Close")
main.plot(x="Days", title="Infosys", y="Line of Best-Fit", ax=ax, color="red")
```
